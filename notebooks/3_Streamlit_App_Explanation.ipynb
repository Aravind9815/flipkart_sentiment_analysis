### Step 1: Import Required Libraries

```python
import streamlit as st
import pickle
import warnings
warnings.filterwarnings("ignore")
```

### Step 2: Load Trained Model and Vectorizer

```python
model = pickle.load(open("sentiment_model.pkl", "rb"))
vectorizer = pickle.load(open("vectorizer.pkl", "rb"))
```

### Observation:
The trained Logistic Regression model and TF-IDF vectorizer

### Observation:
A clear title and description were added to help users

### Step 4: Accept User Input

```python
user_review = st.text_area("Enter your review here")
```

### Observation:
A text area was provided to allow users to input
customer reviews for sentiment analysis.

### Step 5: Predict Sentiment on Button Click

```python
if st.button("Predict Sentiment"):
    review_vector = vectorizer.transform([user_review])
    prediction = model.predict(review_vector)
```

### Observation:
When the user clicks the prediction button, the input review
is transformed using the TF-IDF vectorizer and passed to the model.


### Step 6 : Display Prediction Result

```python
    if prediction[0] == 1:
        st.success("Positive Review 😊")
    else:
        st.error("Negative Review 😞")
```

### Step 7: Handle Empty Input (Good Practice)

```python
if st.button("Predict Sentiment") and user_review.strip() == "":
    st.warning("Please enter a review before predicting sentiment.")
```

### Observation:
Input validation was added to prevent predictions on empty reviews,
improving the robustness of the application.


### Step 8: Complete Streamlit App Code (Final app.py)

```python
import streamlit as st
import pickle

model = pickle.load(open("sentiment_model.pkl", "rb"))
vectorizer = pickle.load(open("vectorizer.pkl", "rb"))

st.title("Sentiment Analysis of Flipkart Product Reviews")
st.write("Enter a customer review below to predict its sentiment.")

user_review = st.text_area("Enter your review here")

if st.button("Predict Sentiment"):
    if user_review.strip() == "":
        st.warning("Please enter a review before predicting sentiment.")
    else:
        review_vector = vectorizer.transform([user_review])
        prediction = model.predict(review_vector)

        if prediction[0] == 1:
            st.success("Positive Review 😊")
        else:
            st.error("Negative Review 😞")
```
